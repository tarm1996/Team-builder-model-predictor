```python
gamestats1718 = pd.read_csv('P1-1718.csv')
gamestats1819 = pd.read_csv('P1-1819.csv')
gamestats1920 = pd.read_csv('P1-1920.csv')
gamestats2021 = pd.read_csv('P1-2021.csv')
```

```python
gamestats1718 = gamestats1718.loc[:, :'AR']
gamestats1819 = gamestats1819.loc[:, :'AR']
gamestats1920 = gamestats1920.loc[:, :'AR']
gamestats2021 = gamestats2021.loc[:, :'AR']
```

```python
gamestats1718 = gamestats1718.drop(['Div', 'FTR', 'HTR'], axis=1)
gamestats1819 = gamestats1819.drop(['Div', 'FTR', 'HTR'], axis=1)
gamestats1920 = gamestats1920.drop(['Div', 'Time', 'FTR', 'HTR'], axis=1)
gamestats2021 = gamestats2021.drop(['Div', 'Time', 'FTR', 'HTR'], axis=1)
```

```python
gamestats1718_tail = gamestats1718[300:]
gamestats1819_tail = gamestats1819[300:]
gamestats1920_tail = gamestats1920[300:]
```

```python
gamestats1718_head = gamestats1718[:300]
gamestats1819_head = gamestats1819[:300]
gamestats1920_head = gamestats1920[:300]
```

```python
DF_gamestats = pd.concat([gamestats1718_head,gamestats1819_head,gamestats1920_head,gamestats2021,gamestats1718_tail,gamestats1819_tail,gamestats1920_tail], axis=0)
DF_gamestats.reset_index(inplace=True)
DF_gamestats.drop(['index'], axis=1,inplace=True)
DF_gamestats
```

```python
DF_gamestats.rename(columns = {'FTHG':'Full Time Home Team Goals','FTAG':'Full Time Away Team Goals','HTHG':'Half Time Home Team Goals','HTAG':'Half Time Away Team Goals','HS':'Home Team Shots','AS':'Away Team Shots','HST':'Home Team Shots on Target','AST':'Away Team Shots on Target','HHW':'Home Team Hit Woodwork','AHW':'Away Team Hit Woodwork','HC':'Home Team Corners','AC':'Away Team Corners','HF':'Home Team Fouls Committed','AF':'Away Team Fouls Committed','HFKC':'Home Team Free Kicks Conceded','AFKC':'Away Team Free Kicks Conceded','HO':'Home Team Offsides','AO':'Away Team Offsides','HY':'Home Team Yellow Cards','AY':'Away Team Yellow Cards','HR':'Home Team Red Cards','AR':'Away Team Red Cards'}, inplace = True)
```

```python
DF_gamestats.drop(['HomeTeam','AwayTeam'], axis=1,inplace=True)
```
